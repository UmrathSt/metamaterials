```python
import numpy as np
from scipy.optimize import minimize
from scipy.optimize import leastsq
from scipy.special import jvp, yvp
from matplotlib import pyplot as plt
import sympy as sy
```

```python
def Zmodel1(f, R, L, C):
    """ impedance of a series RLC circuit
    """
    w = 2*np.pi*f
    return (R + 1j*(w*L - 1/(w*C)))

def Zmodel2(f, R, L, C, Lp, Cp):
    """ impedance of a series RLC circuit
    """
    w = 2*np.pi*f
    return ((R + 1j*(w*L - 1/(w*C)))**(-1) + (1j*(w*Lp - 1/(w*Cp)))**(-1))**(-1)


def residuals(params, Zm, f, Z):
    diff = Zm(f, *params) - Z
    return diff.real**2 + diff.imag**2

def Zslab(f, D, epsR, kappa):
    eps = epsR + 1j*kappa/(2*np.pi*f*8.85e-12)
    return 376.7j/np.sqrt(eps)*np.tan(2*np.pi*f*np.sqrt(eps)*D/3e8)
    
```

```python
R = ["0.1", "1", "10", "100", "200"]#"100", "200", "300"]#np.linspace(2.35, 3.35, 11)
data = [np.loadtxt("TL_Model_RectRing/S11_f_fss_L_8_w_1_Ohm_%s_Cap_no.txt" %r, delimiter=",") for r in R]
adata = [np.loadtxt("TL_Model_RectRing/S11_f_absorber_L_9.5_w_1_Ohm_%s_Cap_no.txt" %r, delimiter=",") for r in R]
f = data[0][:,0]
S11 =  [dat[:,1]+1j*dat[:,2] for dat in data]
S21 =  [dat[:,3]+1j*dat[:,4] for dat in data]
Zfss = [dat[:,5]+1j*dat[:,6] for dat in data]
S11a =  [dat[:,1]+1j*dat[:,2] for dat in adata]
Zfssa = [dat[:,5]+1j*dat[:,6] for dat in adata]
```

```python
fig = plt.figure(figsize=(8,8))
ax1 = fig.add_subplot(311)
ax2 = fig.add_subplot(312)
ax3 = fig.add_subplot(313)
cols = ["r", "b", "g", "m", "c", "y", "k"]
for idx in range(len(R)):
    ax1.plot(data[0][:,0]/1e9, Zfss[idx].real, linestyle="-",   color=cols[idx], label="Re(Z) R=%s" %R[idx])
    ax1.plot(data[0][:,0]/1e9, Zfss[idx].imag, linestyle="--", color=cols[idx], label="Im(Z) R=%s" %R[idx])
    ax1.plot([0,20], [0,0], "k--")
    ax2.plot(data[0][:,0]/1e9, 20*np.log10(abs(S11[idx])), linestyle="-", color=cols[idx], label="S11, R=%s" %R[idx])
    ax2.plot(data[0][:,0]/1e9, 20*np.log10(abs(S21[idx])), linestyle="--", color=cols[idx])
    ax3.plot(data[0][:,0]/1e9, 1-np.abs(S11[idx])**2-np.abs(S21[idx])**2, color=cols[idx])
ax1.set_ylabel("Re/Im Z")
ax2.set_ylabel("|S11| [dB]")
ax2.set_xlabel("f [GHz]")
ax3.set_ylabel("Absorbed Power")
ax1.set_xlim([1,20])
ax2.set_xlim([1,20])
ax2.set_ylim([-40,0.1])
ax2.legend(loc="best").draw_frame(False)
fig.tight_layout()
plt.show()
```
